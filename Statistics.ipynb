1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test for a single sample mean.

    :param sample: List or numpy array of sample data
    :param pop_mean: Known population mean
    :param pop_std: Known population standard deviation
    :param alpha: Significance level (default = 0.05 for 95% confidence)
    :return: Z-score, p-value, and test conclusion
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference between sample and population mean."

    return z_score, p_value, conclusion

# Example usage:
sample_data = [102, 98, 101, 104, 100, 97, 99, 103, 105, 98]
population_mean = 100
population_std_dev = 2.5  # Known population standard deviation

z_score, p_value, result = z_test(sample_data, population_mean, population_std_dev)
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {result}")
```

4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for a single sample mean.

    :param sample: List or numpy array of sample data
    :param pop_mean: Known population mean
    :param pop_std: Known population standard deviation
    :param alpha: Significance level (default = 0.05)
    :return: Z-score, p-value, and test conclusion
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Critical Z-value for two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Interpretation
    if abs(z_score) > z_critical:
        conclusion = "Reject the null hypothesis: Significant difference."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference."

    return z_score, p_value, z_critical, conclusion

# Generate random sample data
np.random.seed(42)
sample_size = 30
population_mean = 100
population_std_dev = 15  # Known population standard deviation
sample_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

# Perform two-tailed Z-test
z_score, p_value, z_critical, result = two_tailed_z_test(sample_data, population_mean, population_std_dev)

# Print results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{z_critical:.2f}")
print(f"Conclusion: {result}")

# Visualization of the decision region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Shade rejection regions (left and right tails)
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color="red", alpha=0.3, label="Rejection Region")

# Shade acceptance region (center)
plt.fill_between(x, y, where=(x >= -z_critical) & (x <= z_critical), color="green", alpha=0.3, label="Acceptance Region")

# Mark critical Z-values
plt.axvline(-z_critical, color='red', linestyle="--", label=f"Z = {-z_critical:.2f}")
plt.axvline(z_critical, color='red', linestyle="--", label=f"Z = {z_critical:.2f}")

# Mark observed Z-score
plt.axvline(z_score, color='black', linestyle="-.", label=f"Observed Z = {z_score:.2f}")

plt.title("Two-Tailed Z-test Decision Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(pop_mean=100, pop_std=15, sample_size=30, alpha=0.05, effect_size=0.5):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    :param pop_mean: Population mean under null hypothesis (H0).
    :param pop_std: Population standard deviation (assumed known).
    :param sample_size: Number of observations in the sample.
    :param alpha: Significance level (default 0.05 for 95% confidence).
    :param effect_size: Difference between true mean (H1) and null mean, measured in standard deviations.
    """

    # Compute standard error
    std_error = pop_std / np.sqrt(sample_size)

    # Null hypothesis distribution (H0)
    x = np.linspace(pop_mean - 4*pop_std, pop_mean + 4*pop_std, 1000)
    h0_pdf = stats.norm.pdf(x, pop_mean, std_error)

    # Alternative hypothesis distribution (H1)
    true_mean = pop_mean + (effect_size * pop_std)  # True mean under H1
    h1_pdf = stats.norm.pdf(x, true_mean, std_error)

    # Critical z-value for two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)
    critical_value_right = pop_mean + z_critical * std_error
    critical_value_left = pop_mean - z_critical * std_error

    # Visualization
    plt.figure(figsize=(10, 6))

    # Plot H0 and H1 distributions
    plt.plot(x, h0_pdf, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, h1_pdf, label="Alternative Hypothesis (H1)", color="green")

    # Shade Type 1 error region (α) - Rejection of true H0
    plt.fill_between(x, h0_pdf, where=(x > critical_value_right) | (x < critical_value_left), color="red", alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 error region (β) - Failure to reject false H0
    beta_region = (x > critical_value_left) & (x < critical_value_right)
    plt.fill_between(x, h1_pdf, where=beta_region, color="purple", alpha=0.3, label="Type 2 Error (β)")

    # Vertical lines for critical values
    plt.axvline(critical_value_right, color='red', linestyle="--", label=f"Critical Value (+{z_critical:.2f})")
    plt.axvline(critical_value_left, color='red', linestyle="--", label=f"Critical Value (-{z_critical:.2f})")

    # Labels and legend
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

# Example usage:
visualize_type1_type2_errors()
```

6 Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent (two-sample) T-test.

    :param sample1: First sample data (numpy array or list)
    :param sample2: Second sample data (numpy array or list)
    :param alpha: Significance level (default = 0.05)
    :return: t-statistic, p-value, and test conclusion
    """
    # Perform independent T-test (equal variance assumed by default)
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: The two group means are significantly different."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference between the two group means."

    return t_stat, p_value, conclusion

# Generate random sample data for two independent groups
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)  # Sample 1 (Mean = 100, Std = 15)
group2 = np.random.normal(loc=105, scale=15, size=30)  # Sample 2 (Mean = 105, Std = 15)

# Perform T-test
t_stat, p_value, result = independent_t_test(group1, group2)

# Print results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {result}")

```

7 Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    """
    Perform a paired sample T-test and visualize the results.

    :param before: Pre-treatment measurements (numpy array or list)
    :param after: Post-treatment measurements (numpy array or list)
    :param alpha: Significance level (default = 0.05)
    :return: T-statistic, p-value, and test conclusion
    """
    # Compute paired t-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: There is a significant difference."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference."

    # Visualization
    plt.figure(figsize=(8, 5))

    # Plot before and after values
    plt.scatter(range(len(before)), before, color='blue', label="Before Treatment", alpha=0.6)
    plt.scatter(range(len(after)), after, color='green', label="After Treatment", alpha=0.6)

    # Connect paired points with lines
    for i in range(len(before)):
        plt.plot([i, i], [before[i], after[i]], color='gray', alpha=0.5)

    plt.title("Paired Sample Comparison: Before vs. After")
    plt.xlabel("Subject Index")
    plt.ylabel("Measurement Value")
    plt.legend()
    plt.show()

    return t_stat, p_value, conclusion

# Generate simulated before-and-after data
np.random.seed(42)
before_treatment = np.random.normal(loc=100, scale=10, size=30)  # Pre-treatment measurements
after_treatment = before_treatment + np.random.normal(loc=-2, scale=5, size=30)  # Post-treatment with slight decrease

# Perform paired T-test
t_stat, p_value, result = paired_t_test(before_treatment, after_treatment)

# Print results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {result}")
```

```python
import numpy as np
import scipy.stats as stats

def perform_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    std_error = pop_std / np.sqrt(sample_size)  # Standard error

    # Compute Z-score
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

    conclusion = "Reject H0" if p_value < alpha else "Fail to reject H0"
    return z_stat, p_value, conclusion

def perform_t_test(sample, pop_mean, alpha=0.05):
    """
    Perform a one-sample T-test.
    """
    t_stat, p_value = stats.ttest_1samp(sample, pop_mean)
    conclusion = "Reject H0" if p_value < alpha else "Fail to reject H0"
    return t_stat, p_value, conclusion

# Simulate data (Sample from normal distribution)
np.random.seed(42)
sample_size = 25  # Small sample (for T-test)
pop_mean = 100  # Known population mean
pop_std = 15  # Known population std dev (for Z-test)

# Generate a random sample
sample = np.random.normal(loc=102, scale=pop_std, size=sample_size)

# Perform Z-test (only valid if population std dev is known)
z_stat, z_p_value, z_conclusion = perform_z_test(sample, pop_mean, pop_std)

# Perform T-test (more appropriate for small samples)
t_stat, t_p_value, t_conclusion = perform_t_test(sample, pop_mean)

# Print results
print(f"Z-Test Results:\n  Z-statistic: {z_stat:.2f}, P-value: {z_p_value:.4f}, Conclusion: {z_conclusion}\n")
print(f"T-Test Results:\n  T-statistic: {t_stat:.2f}, P-value: {t_p_value:.4f}, Conclusion: {t_conclusion}")
```

9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95, pop_std=None):
    """
    Calculate the confidence interval for a sample mean.

    :param sample: List or numpy array of sample data.
    :param confidence: Confidence level (default 0.95 for 95% CI).
    :param pop_std: Population standard deviation (if known, use Z-distribution; otherwise, use T-distribution).
    :return: Tuple (lower bound, upper bound) of confidence interval.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (Bessel's correction)

    if pop_std:
        # Use Z-distribution if population std dev is known
        z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)
        margin_of_error = z_critical * (pop_std / np.sqrt(sample_size))
    else:
        # Use T-distribution if population std dev is unknown
        t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df=sample_size - 1)
        margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Sample from normal distribution

# Compute confidence interval (assuming unknown population standard deviation)
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95, pop_std=None):
    """
    Calculate the margin of error for a given confidence level using sample data.

    :param sample: List or numpy array of sample data.
    :param confidence: Confidence level (default is 0.95 for 95% confidence).
    :param pop_std: Population standard deviation (if known, use Z-distribution; otherwise, use T-distribution).
    :return: Margin of error (MoE)
    """
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (Bessel's correction)

    if pop_std:
        # Use Z-distribution if population standard deviation is known
        z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)
        margin_error = z_critical * (pop_std / np.sqrt(sample_size))
    else:
        # Use T-distribution if population standard deviation is unknown
        t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df=sample_size - 1)
        margin_error = t_critical * (sample_std / np.sqrt(sample_size))

    return margin_error

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data

# Compute margin of error (assuming unknown population standard deviation)
moe = margin_of_error(sample_data)
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, sensitivity, false_positive):
    """
    Compute the posterior probability using Bayes' Theorem.

    :param prior: Prior probability of disease (P(H))
    :param sensitivity: True positive rate (P(E|H))
    :param false_positive: False positive rate (P(E|~H))
    :return: Posterior probability (P(H|E))
    """
    complement_prior = 1 - prior  # Probability of not having the disease P(~H)

    # Total probability of testing positive: P(E) = P(E|H)P(H) + P(E|~H)P(~H)
    marginal_prob = (sensitivity * prior) + (false_positive * complement_prior)

    # Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (sensitivity * prior) / marginal_prob

    return posterior

# Given values:
prior_prob = 0.01  # 1% have the disease
true_positive_rate = 0.90  # 90% test correctly
false_positive_rate = 0.05  # 5% false positives

# Compute posterior probability
posterior_prob = bayes_theorem(prior_prob, true_positive_rate, false_positive_rate)

print(f"Probability of having the disease given a positive test: {posterior_prob:.4f}")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

def compute_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    :param observed: 2D array (contingency table) of observed values.
    :return: Expected frequency table (numpy array).
    """
    observed = np.array(observed)  # Convert to NumPy array
    row_totals = observed.sum(axis=1).reshape(-1, 1)  # Sum of rows
    col_totals = observed.sum(axis=0).reshape(1, -1)  # Sum of columns
    grand_total = observed.sum()  # Total sum of all elements

    # Compute expected frequencies
    expected = (row_totals @ col_totals) / grand_total  # Matrix multiplication

    return expected

# Example Observed Data (Contingency Table)
observed_data = np.array([[50, 30],  # Males (Bought, Not Bought)
                          [20, 40]]) # Females (Bought, Not Bought)

# Compute Expected Frequencies
expected_frequencies = compute_expected_frequencies(observed_data)

# Convert to DataFrame for better readability
df_expected = pd.DataFrame(np.round(expected_frequencies, 2),
                           columns=["Bought", "Not Bought"],
                           index=["Male", "Female"])

# Print Expected Frequencies
print("Expected Frequencies:")
print(df_expected)
```

16 Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

# Generate two random samples from normal distributions
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15, n=30

# Calculate sample variances
var1 = np.var(sample1, ddof=1)  # Unbiased estimate (ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic (larger variance / smaller variance)
if var1 > var2:
    F_stat = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom
else:
    F_stat = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Compute p-value (two-tailed test)
p_value = 2 * min(stats.f.cdf(F_stat, dfn, dfd), 1 - stats.f.cdf(F_stat, dfn, dfd))

# Print results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject H0 → Variances are significantly different.")
else:
    print("\nConclusion: Fail to Reject H0 → No significant difference in variances.")
```

19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups):
    """
    Checks normality and homogeneity of variances for ANOVA assumptions.

    Parameters:
    groups (arrays): Multiple groups of sample data.

    Returns:
    None (Prints test results)
    """

    # Normality Test (Shapiro-Wilk)
    print("### Normality Test (Shapiro-Wilk) ###")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: W={stat:.4f}, p-value={p:.4f}")
        if p < 0.05:
            print(f"  → Group {i+1} is NOT normally distributed (p < 0.05).")
        else:
            print(f"  → Group {i+1} is normally distributed (p >= 0.05).")

    # Homogeneity of Variances Test (Levene's Test)
    print("\n### Homogeneity of Variance Test (Levene) ###")
    stat, p = stats.levene(*groups)
    print(f"Levene’s Test: W={stat:.4f}, p-value={p:.4f}")
    if p < 0.05:
        print("  → Variances are NOT equal (p < 0.05). ANOVA may not be valid.")
    else:
        print("  → Variances are equal (p >= 0.05).")

    print("\n### Independence Assumption ###")
    print("  - Ensure data is collected independently (e.g., random sampling).")

# Example Usage
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Normal distribution
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Run the assumption checks
check_anova_assumptions(group1, group2, group3)
```

```python
import numpy as np
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# Generate random normal data (Sample)
sample_size = 30
true_variance = 25  # Actual variance of the population
sample = np.random.normal(loc=50, scale=np.sqrt(true_variance), size=sample_size)  # Std Dev = sqrt(variance)

# Hypothesized variance
sigma_0_squared = 20  # Hypothesized variance

# Calculate sample variance
sample_variance = np.var(sample, ddof=1)  # Unbiased estimator (ddof=1)

# Compute Chi-square statistic
chi_square_stat = (sample_size - 1) * sample_variance / sigma_0_squared

# Compute p-value (Two-tailed test)
p_value = 2 * min(stats.chi2.sf(chi_square_stat, df=sample_size-1), stats.chi2.cdf(chi_square_stat, df=sample_size-1))

# Significance level
alpha = 0.05

# Print results
print("\n### Variance Hypothesis Test Results ###")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Significant difference detected. Reject H0.")
else:
    print("No significant difference detected. Fail to reject H0.")
```

25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

# Sample data: Number of successes and sample sizes
n1, x1 = 200, 60  # Group 1: 60 successes out of 200
n2, x2 = 180, 40  # Group 2: 40 successes out of 180

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
p_pool = (x1 + x2) / (n1 + n2)

# Standard error calculation
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Compute Z-score
z_score = (p1 - p2) / se

# Compute two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Print results
print("\n### Two-Proportion Z-Test Results ###")
print(f"Proportion 1 (Group 1): {p1:.4f}")
print(f"Proportion 2 (Group 2): {p2:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Significant difference detected. Reject H0.")
else:
    print("No significant difference detected. Fail to reject H0.")
```

26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate two random normal datasets with different variances
n1, n2 = 30, 30  # Sample sizes
data1 = np.random.normal(loc=50, scale=10, size=n1)  # Std = 10
data2 = np.random.normal(loc=50, scale=15, size=n2)  # Std = 15

# Compute sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Compute F-statistic (ensure F >= 1)
if var1 > var2:
    F_stat = var1 / var2
    dfn, dfd = n1 - 1, n2 - 1
else:
    F_stat = var2 / var1
    dfn, dfd = n2 - 1, n1 - 1

# Compute two-tailed p-value
p_value = 2 * min(stats.f.sf(F_stat, dfn, dfd), stats.f.cdf(F_stat, dfn, dfd))

# Significance level
alpha = 0.05

# Print results
print("\n### F-Test for Variance Comparison ###")
print(f"Variance 1: {var1:.4f}")
print(f"Variance 2: {var2:.4f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Significant difference detected. Reject H0.")
else:
    print("No significant difference detected. Fail to reject H0.")

# Visualization: Boxplot of the two datasets
plt.figure(figsize=(8, 5))
sns.boxplot(data=[data1, data2], palette="Set2")
plt.xticks([0, 1], ["Group 1", "Group 2"])
plt.title("Comparison of Variances Using Boxplot")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

27  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulated observed data (e.g., survey responses from different categories)
observed = np.array([50, 30, 20, 25, 15])  # Observed counts

# Expected frequencies based on a hypothesized distribution
total_observed = np.sum(observed)
expected_ratios = np.array([0.30, 0.25, 0.20, 0.15, 0.10])  # Expected proportions
expected = total_observed * expected_ratios  # Convert to expected counts

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print("\n### Chi-Square Goodness-of-Fit Test Results ###")
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Significant difference detected. Reject H0.")
else:
    print("No significant difference detected. Fail to reject H0.")

# Visualization: Bar plot comparing observed vs. expected frequencies
categories = ["A", "B", "C", "D", "E"]
x = np.arange(len(categories))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, observed, width=0.4, label="Observed", color="blue", alpha=0.7)
plt.bar(x + 0.2, expected, width=0.4, label="Expected", color="red", alpha=0.7)
plt.xticks(x, categories)
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Chi-Square Goodness-of-Fit: Observed vs. Expected")
plt.legend()
plt.show()
```
